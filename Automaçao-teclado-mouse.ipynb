# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/u/1/folders/13D9zOYXuaqRlwMo9PvOtq4vdJHLvOv5r

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time
import pyperclip
pyautogui.PAUSE = 1
pyautogui.alert('Aperte "ok" e não mecha mais.')

#passo 1: abrir uma nova aba
pyautogui.hotkey('ctrl', 't')

#passo 2: copiar e colar o link e apertar na pagina
link = 'https://drive.google.com/drive/u/1/folders/13D9zOYXuaqRlwMo9PvOtq4vdJHLvOv5r'
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(7)

#passo 3: entrar na pagina exporta
pyautogui.click(385, 309, clicks=2)
time.sleep(5)

#passo 4: fazer download do arquivo
pyautogui.click(366, 412)
pyautogui.click(1160, 192)

pyautogui.click(908, 599)
time.sleep(10)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas as pd

tabela = pd.read_excel(r"C:\Users\Daniel Paixão\Desktop\Python\Vendas - Dez.xlsx")
Faturamento = tabela['Valor Final'].sum() #soma da coluna valor final
Quantidade = tabela['Quantidade'].sum() #soma da coluna quantidade
display(tabela)
print(Faturamento)
print(Quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
#passo 5: abrir uma nova aba
pyautogui.hotkey('ctrl', 't')

#passo 6: entrar no link do gmail
gmail = 'https://mail.google.com/mail/u/1/?zx=o2b7l9psevjs#inbox'
pyperclip.copy(gmail)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(10)

#passo 7: clicar para escrever o gmail
pyautogui.click(106, 208)

#passo 8: para quem vai o gmail
pyautogui.write('diogeneskauam1+diretoria@gmail.com')
pyautogui.press('tab') #selecionar o gmail
pyautogui.press('tab') #passar para o campo de assunto

#passo 9: assunto do gmail
assunto = 'relatório de vendas'
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

pyautogui.press('tab') #passar para o campo de escrever gmail

#passo 10: escrever o gmail
texto_gmail = f"""Prezados, bom dia
O faturamento de ontem foi de: R$ {Faturamento:,.2f}
A quantidade de produtos foi de: {Quantidade:,}

Abs
Kauam"""

pyautogui.write(texto_gmail)

#passo 11: apertar em "enviar"
pyautogui.hotkey('ctrl', 'enter')
```
